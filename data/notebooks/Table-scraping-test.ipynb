```python
# Als einzelne csv Dateien speichern

for i, table in enumerate(tables): # enumerate()
    df_table = pd.read_html(StringIO(str(table)))[0] #str notwendig um die Tabelle an pandas zu übergeben, [0] notwendig da read_html() ein list[DataFrame] zurückliefert. StringIO -> read.html(str) ist deprecated und wird in der Zukunft entfernt. Daher in StringIO wrappen
    df_table.to_csv(f"table_{i}.csv", index=False)
    print(df_table)

```

```python
# Als Arbeitsmappen in einer Excel-Datei speichern

with pd.ExcelWriter("Arbeitsmappen.xlsx", engine="openpyxl") as writer:

    for i, table in enumerate(tables):
        table_df = pd.read_html(StringIO(str(table)))[0]
        table_df.to_excel(writer,sheet_name=f"Sheet_{i}")

```
