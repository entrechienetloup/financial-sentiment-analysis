```python
from bs4 import BeautifulSoup
import requests
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
from io import StringIO

driver = webdriver.Chrome()
url = "https://de.wikipedia.org/wiki/TecDAX"
driver.get(url)
```

```python
soup = BeautifulSoup(driver.page_source, "html.parser")
print(soup)
```

```python
tables = soup.find(id="zusammensetzung")
print(tables)
```

```python
html_string = str(tables)
html_io = StringIO(html_string)
df_list = pd.read_html(html_io, decimal=',', thousands='.')

df = df_list[0]

if 'Logo' in df.columns:
    df = df.drop(columns=['Logo'])

print(df.head())
```

```python
namen_liste_unique = df['Name'].unique().tolist()
driver.quit()
```
